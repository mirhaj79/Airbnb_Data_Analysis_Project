# ðŸ”µData Overview
My case study focuses on Airbnb data in New York City, covering the period until October 2023. New York City, known for its iconic skyline and diverse neighborhoods, is a bustling metropolis on the northeastern coast of the United States. The dataset was obtained from ' www.insideairbnb.com '.
<br>
<br>The dataset have 75 variables, and among them, I've provided concise descriptions for 40 that were useful for my project:
<br>
<br>**Host Information:**
<br> 1. listing_url: Unique url for the host.
<br> 2. host_id: Unique identifier for the host.
<br> 3. host_name: Name of the host.
<br> 4. host_since: The date when the host joined Airbnb.
<br> 5. host_response_rate: The response rate of the host to guest inquiries.
<br> 6. host_acceptance_rate: The rate at which the host accepts booking requests.
<br> 7. calculated_host_listings_count: The total number of listings by the host.
<br> 8. calculated_host_listings_count_entire_homes: Count of entire homes/apartments listed by the host.
<br> 9. calculated_host_listings_count_private_rooms: Count of private rooms listed by the host.
<br> 10. calculated_host_listings_count_shared_rooms: Count of shared rooms listed by the host.
<br>
<br>**Location Information:**
<br> 11. neighbourhood_cleansed: The name of the neighborhood where the listing is located.
<br> 12. neighbourhood_group_cleansed: The name of the neighborhood group or region.
<br> 13. latitude: Latitude coordinates of the listing.
<br> 14. longitude: Longitude coordinates of the listing.
<br>
<br>**Listing Details:**
<br> 15. room_type: Type of room (e.g., Entire home/apartment, Private room, Shared room).
<br> 16. accommodates: The number of guests the listing can accommodate.
<br> 17. bathrooms: The number of bathrooms in the listing.
<br> 18. bathrooms_text: A textual description of the bathrooms.
<br> 19. bedrooms: The number of bedrooms in the listing.
<br> 20. beds: The number of beds available in the listing.
<br> 21. amenities: A list of amenities provided with the listing.
<br>
<br>**Pricing and Availability:**
<br> 22. price: The price of the listing.
<br> 23. minimum_nights: The minimum number of nights required for booking.
<br> 24. maximum_nights: The maximum number of nights a guest can stay.
<br> 25. availability_30: Number of available nights in the next 30 days.
<br> 26. availability_60: Number of available nights in the next 60 days.
<br> 27. availability_90: Number of available nights in the next 90 days.
<br> 28. availability_365: Number of available nights in the next 365 days.
<br>
<br>**Reviews and Ratings:**
<br> 29. number_of_reviews: Total number of reviews for the listing.
<br> 30. number_of_reviews_ltm: Number of reviews in the last 12 months.
<br> 31. number_of_reviews_l30d: Number of reviews in the last 30 days.
<br> 32. first_review: When first review have been written
<br> 33. reviews_per_month: Average number of reviews per month for the listing.
<br> 34. review_scores_rating: Overall rating of the listing.
<br> 35. review_scores_accuracy: Rating for accuracy in the listing description.
<br> 36. review_scores_cleanliness: Rating for cleanliness.
<br> 37. review_scores_checkin: Rating for the check-in process.
<br> 38. review_scores_communication: Rating for communication with the host.
<br> 39. review_scores_location: Rating for the location of the listing.
<br> 40. review_scores_value: Rating for the value provided.

# ðŸ”µData Wrangling

```python
# Importing essential libraries for data analysis and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from the specified file path
df = pd.read_csv('C:/Users/MohammadAmin/Desktop/Inside Airbnb PROJECT/data/df.csv')
```

```python
# Display the first few rows of the dataset to inspect its structure
print(df.head())
```

```python
# Print the column names of the dataset to list the available variables
print(df.columns)
```

```python
# List of columns to keep
columns_to_keep = [
    'listing_url', 'host_id', 'host_name', 'host_since',
    'neighbourhood_cleansed', 'neighbourhood_group_cleansed', 'latitude', 'longitude',
    'room_type', 'accommodates', 'bathrooms_text', 'bedrooms', 'beds', 'amenities',
    'price', 'minimum_nights', 'maximum_nights', 'availability_30', 'availability_60', 'availability_90',
    'availability_365', 'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d',
    'first_review', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value',
    'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes',
    'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms',
    'reviews_per_month'
]

# Create a new DataFrame with only the specified columns
df = df[columns_to_keep]
```

```python
# Print the column names of the dataset to list the keeped variables
print(df.columns)
```

```python
# Display the shape of the dataset to show the number of rows and columns
print(df.shape)
```

```python
# Show the data types of each column in the dataset
print(df.dtypes)
```

```python
# Remove non-numeric characters from the "price" column
df['price'] = df['price'].str.replace('$', '').str.replace(',', '')

# Convert "price" to float, handling potential errors by using errors='coerce'
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Replace any NaN values with a default value (e.g., 0)
df['price'].fillna(0, inplace=True)

# Print the "price" column
print(df['price'])
```

```python
# Create a missing data matrix (True for missing data, False for non-missing)
missing_data = df.isnull()

# Create a heatmap of the missing data
plt.figure(figsize=(12, 6))  # Set the size of the heatmap
sns.heatmap(missing_data, cbar=False, cmap='viridis')

# Customize the plot
plt.title('Missing Data Heatmap')
plt.xlabel('Columns')
plt.ylabel('Rows')

# Display the plot
plt.show()
```

```python
# Retrieve detailed information about the dataset
print(df.info())
```

```python
# Count missing values in each column
missing_count = df.isna().sum()

# Display the missing value counts
print(missing_count)
```

```python
# Fill missing values in the "host_name" column with "Unknown"
df['host_name'].fillna('Unknown', inplace=True)

# Fill missing values in the "host_since" column with "no date"
df['host_since'].fillna('no date', inplace=True)

# Fill missing values in the "bathrooms_text" column with "no data"
df['bathrooms_text'].fillna('no data', inplace=True)

# Calculate the median of the "bedrooms" column
median_bedrooms = df['bedrooms'].median()
# Fill missing values in the "bedrooms" column with the median
df['bedrooms'].fillna(median_bedrooms, inplace=True)

# Calculate the median of the "beds" column
median_beds = df['beds'].median()
# Fill missing values in the "beds" column with the median
df['beds'].fillna(median_beds, inplace=True)

# Fill missing values in the "first_review" column with 0
df['first_review'].fillna(0, inplace=True)

# Fill missing values in the "reviews_per_month" column with 0
df['reviews_per_month'].fillna(0, inplace=True)

# List of columns to fill with "no score"
columns_to_fill = [
    'review_scores_rating',
    'review_scores_accuracy',
    'review_scores_cleanliness',
    'review_scores_checkin',
    'review_scores_communication',
    'review_scores_location',
    'review_scores_value'
]

# Fill missing values in the specified columns with "no score"
for column in columns_to_fill:
    df[column].fillna('no score', inplace=True)
```

```python
# Count missing values in each column
missing_count = df.isna().sum()

# Display the missing value counts
print(missing_count)
```

```python
# Check for duplicate rows in the DataFrame
duplicate_rows = df[df.duplicated(keep='first')]

# Display the duplicate rows, if any
print(duplicate_rows)
```

```python
# Generate descriptive statistics of the dataset
print(df.describe())
```

```python
# Create a boxplot for the "price" column
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'])
plt.title('Boxplot for Price')
plt.xlabel('Price')
plt.show()
```

```python
# Calculate the correlation matrix for numeric columns in the dataset, and generate a heatmap
correlation_matrix = df.corr(numeric_only=True)

# Create a mask to hide the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create a heatmap with the mask
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", mask=mask)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Create a violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='neighbourhood_group_cleansed', y='availability_365', data=df, palette='Set2')
plt.title('Relation between Neighbourhood Group and Availability (Violin Plot)')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Availability in 365 Days')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Calculate the average price for each neighborhood group
average_prices = df.groupby('neighbourhood_group_cleansed')['price'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='neighbourhood_group_cleansed', y='price', data=average_prices, palette='Set2')
plt.title('Average Price by Neighbourhood Group Cleansed (Bar Plot)')
plt.xlabel('Neighbourhood Group Cleansed')
plt.ylabel('Average Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Annotate the bars with average price values
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# Define colors for each neighbourhood group
colors = {'Bronx': 'red', 'Brooklyn': 'blue', 'Manhattan': 'green', 'Queens': 'orange', 'Staten Island': 'purple'}

# Create a scatterplot with colors based on neighbourhood group
plt.figure(figsize=(12, 10))
for group, color in colors.items():
    listings = df[df['neighbourhood_group_cleansed'] == group]
    plt.scatter(listings['longitude'], listings['latitude'], s=15, alpha=0.6, color=color, label=group)

plt.title('Airbnb Listings Scatterplot by Neighbourhood Group')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Adding a legend
legend = plt.legend(title='Neighbourhood Group', loc='upper left')
for handle in legend.legend_handles:
    handle.set_sizes([30])

# Adding a grid
plt.grid(True, linestyle='--', alpha=0.4)

# Adjusting axis limits for better display
plt.xlim(df['longitude'].min() - 0.02, df['longitude'].max() + 0.02)
plt.ylim(df['latitude'].min() - 0.02, df['latitude'].max() + 0.02)

plt.show()
```

```python
# Assuming df is your DataFrame containing the data

# Set the Seaborn theme to whitegrid for a cleaner look
sns.set_theme(style='whitegrid')

# Create a figure with a custom size
plt.figure(figsize=(12, 8))

# Calculate the average price for each unique latitude and longitude combination
average_price_by_location = df.groupby(['latitude', 'longitude'])['price'].mean().reset_index()

# Filter the data to only include average prices under $11,000
average_price_by_location = average_price_by_location[average_price_by_location['price'] < 1000]

# Create a scatterplot using seaborn with a gradient color palette
sns.scatterplot(data=average_price_by_location, x='longitude', y='latitude', hue='price', palette='viridis', size='price', sizes=(10, 200))

# Add labels to the x-axis and y-axis
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Latitude', fontsize=14)

# Add a title to the plot
plt.title('Average Price by Location under 1000$ (Scatterplot)', fontsize=15)

# Show the plot
plt.show()
```

```python
# Create a scatter plot of "number_of_reviews" vs "price"
plt.figure(figsize=(10, 6))
plt.scatter(df['number_of_reviews'], df['price'], alpha=0.5, color='b')
plt.title('Number of Reviews vs Price')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
# Count the occurrences of each room type
room_type_counts = df['room_type'].value_counts()

# Calculate the percentage for each room type
total = len(df)
percentages = (room_type_counts / total) * 100

# Create a bar plot
plt.figure(figsize=(8, 6))
barplot = sns.barplot(x=room_type_counts.index, y=room_type_counts.values, palette='Set2')
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Add the percentage on top of each bar
for index, value in enumerate(percentages):
    plt.text(index, room_type_counts.values[index], f'{value:.1f}%', ha='center', va='bottom')

plt.show()
```

```python
# Set the Seaborn theme
sns.set_theme(style='whitegrid')

# Create a figure with a custom size
plt.figure(figsize=(12, 8))

# Create a stripplot with room_type on the x-axis, reviews_per_month on the y-axis, and color by neighbourhood_group_cleansed
sns.stripplot(data=df, x='room_type', y='reviews_per_month', hue='neighbourhood_group_cleansed', jitter=True, dodge=True)

# Add labels to the x-axis and y-axis
plt.xlabel('Room Type', fontsize=14)
plt.ylabel('Reviews per Month', fontsize=14)

# Add a title to the plot
plt.title('Stripplot of Reviews per Month by Room Type', fontsize=15)

# Show the plot with a legend
plt.legend(title='Neighbourhood Group', loc='upper right')
plt.show()
```

```python
# Assuming df is your DataFrame containing the data

# Sum the three columns to get the total calculated host listings count
df['total_calculated_host_listings_count'] = df['calculated_host_listings_count_entire_homes'] + \
                                              df['calculated_host_listings_count_private_rooms'] + \
                                              df['calculated_host_listings_count_shared_rooms']

# Group by 'host_id' and aggregate the maximum 'host_name' and 'total_calculated_host_listings_count' values
grouped_data = df.groupby('host_id').agg({'host_name': 'max', 'total_calculated_host_listings_count': 'max'})

# Sort the hosts by the total calculated host listings count in descending order
sorted_hosts = grouped_data.sort_values(by='total_calculated_host_listings_count', ascending=False)

# Select the top 10 unique host IDs
top_10_unique_hosts = sorted_hosts.head(10)

# Print the top 10 unique host IDs, along with their host_name and total calculated host listings count
for host_id, row in top_10_unique_hosts.iterrows():
    host_name = row['host_name']
    total_count = row['total_calculated_host_listings_count']
    print(f'Host ID: {host_id}, Host Name: {host_name}, Total Count: {total_count}')
```
